### F.1 New Feature "\<fg_percent\>"


```python
# <Student to fill this section>
FEATURE_NAME = "fg_percent"
```

```python
def safe_ratio(numer, denom, eps=1e-6):
    return np.where(denom == 0, 0, numer / (denom + eps))
```

```python
fgm_train = X_train_final["twoPM"] + X_train_final["TPM"]
fga_train = X_train_final["twoPA"] + X_train_final["TPA"]

fgm_test  = X_test_final["twoPM"] + X_test_final["TPM"]
fga_test  = X_test_final["twoPA"] + X_test_final["TPA"]

# Calculating hit rate
X_train_final[FEATURE_NAME] = safe_ratio(fgm_train, fga_train)
X_test_final[FEATURE_NAME]  = safe_ratio(fgm_test,  fga_test)
```

```python
print("train range:", float(X_train_final[FEATURE_NAME].min()), "→", float(X_train_final[FEATURE_NAME].max()))
print("test  range:", float(X_test_final[FEATURE_NAME].min()),  "→", float(X_test_final[FEATURE_NAME].max()))
```

```python
# <Student to fill this section>
feature_engineering_1_explanations = """
We do not have direct FGM/FGA columns, but we do have two-point and three-point makes/attempts (twoPM, twoPA, TPM, TPA). We reconstruct overall field goals made and attempted as FGM = twoPM + TPM and FGA = twoPA + TPA,
and then compute fg_percent = FGM / FGA with a safe division to handle zero attempts. This feature captures overall finishing efficiency and is often predictive of scoring/impact.
"""
```

```python

def split_dataset(X, y, test_size=0.2, random_state=42, stratify=True):
    """
    Parameters:
        X : DataFrame features
        y : Series labels
        test_size : float Test set proportion, default 0.2
        random_state : int Random seed for reproducibility
        stratify : bool Whether to perform stratified sampling by category

    return：
        X_train, X_valid, y_train, y_valid
    """
    if stratify:
        X_train, X_valid, y_train, y_valid = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )
    else:
        X_train, X_valid, y_train, y_valid = train_test_split(
            X, y, test_size=test_size, random_state=random_state
        )

    return X_train, X_valid, y_train, y_valid
```

### G.2 Data Transformation "\<feature filter\>"



```python
# <Student to fill this section>
def drop_leakage_features(X_train, X_test):
    """
    Remove features that may cause information leakage or overfitting
    """
    leakage_features = [
        "player_id_RARE", "Rec_Rank", "Rec_Rank__is_missing",

    ]

    features_to_drop = [col for col in leakage_features if col in X_train.columns]

    print("Dropping features:", features_to_drop)

    X_train_clean = X_train.drop(columns=features_to_drop, errors="ignore")
    X_test_clean = X_test.drop(columns=features_to_drop, errors="ignore")

    return X_train_clean, X_test_clean


X_train_final, X_test_final = drop_leakage_features(X_train_final, X_test_final)
```

```python
print("New train shape:", X_train_final.shape)
print("New test shape:", X_test_final.shape)
```

```python
# <Student to fill this section>
data_transformation_1_explanations = """
We remove features that could cause information leakage or unnecessary overfitting:

• player_id_RARE:
  This is a proxy for a specific identity bucket rather than a skill-related signal.
  Keeping identity-like columns in a small/imbalanced dataset often memorizes players
  instead of learning generalizable patterns.

• Rec_Rank and Rec_Rank__is_missing:
  Recruiting rank is a future-facing or externally curated ranking that is highly
  correlated with the target. It can act as a shortcut feature that leaks label
  information and inflates validation scores. Dropping them prevents target leakage.

Impact:
After dropping these columns, the feature count reduced from 195 to 193.
This improves model robustness and helps XGBoost focus on skill/box-score features
instead of identity or leakage-prone signals. The transformation does not require
normalization/standardization because XGBoost is a tree-based learner and is invariant
to monotonic scaling.
"""
```

```python
# <Student to fill this section>
performance_metrics_explanations = """
In this experiment, we employ multiple metrics to comprehensively evaluate XGBoost's classification performance:

Accuracy
Measures the overall proportion of correct predictions, providing an intuitive reflection of general performance. However, when category distributions are imbalanced, relying solely on accuracy may “mask” the model's ability to identify minority classes (selected players).

Precision
The proportion of players correctly identified as “selected” among all players classified as ‘selected’ by the model. High precision indicates fewer false positives, reducing the risk of misclassifying players who should not be selected as “likely to be selected” (minimizing misallocation of resources).

Recall
The proportion of players correctly identified as “selected” among all players who were actually selected. A high Recall indicates fewer false negatives, minimizing the risk of overlooking genuinely promising players—particularly critical in draft scenarios.

F1-score
The harmonic mean of Precision and Recall. It balances both metrics, serving as a core composite indicator in scenarios requiring control over both “false positives” and “false negatives.”

ROC-AUC (Area Under the Curve)
Measures the model's overall discrimination capability across all thresholds. A value closer to 1 indicates the model better distinguishes between “selected” and “not selected” categories. AUC is more stable with class imbalance, making it highly valuable as a global capability metric.

In summary, we do not rely solely on a single metric but comprehensively evaluate Accuracy, Precision, Recall, F1-score, and AUC. Among these, F1-score and AUC serve as primary decision-making criteria: the former balances false positives and false negatives, while the latter reflects overall discrimination capability. Precision and Recall help me monitor the risks of “over-selection” and “under-selection” from a business perspective, respectively. This approach enables a more comprehensive and robust evaluation of XGBoost's actual performance in talent selection prediction tasks.
"""
```

```python
# <Student to fill this section>
scale_pos_weight = (len(y_train_split) - sum(y_train_split)) / sum(y_train_split)
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    n_estimators=1000,
    max_depth=5,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_lambda=1.0,
    reg_alpha=0.0,
    scale_pos_weight=scale_pos_weight,
    tree_method='hist',
    n_jobs=-1,
    random_state=42
)
```

```python
# <Student to fill this section>
hyperparameters_selection_explanations = """
We tuned the hyperparameters of XGBoost to balance model complexity, generalization, and computational efficiency.
The key parameters chosen are:

- n_estimators=1000 This sets the number of boosting rounds. A moderately large value ensures enough iterations
  for learning, while avoiding excessive overfitting or training time. 

- max_depth=5: Controls the maximum depth of each tree. A depth of 5 is often a good balance, allowing the model
  to capture non-linear interactions without overfitting too heavily on the training data.

- learning_rate=0.03: The step size shrinkage applied at each boosting step. A smaller value makes the model
  learn more gradually, which generally improves generalization. This is paired with a sufficient number of trees
  (n_estimators) to maintain predictive power.

- subsample=0.8 and colsample_bytree=0.8: These sampling parameters introduce randomness during training by using
  only a fraction of the rows and features for each tree. This reduces variance and helps prevent overfitting.

- reg_lambda=1.0 and reg_alpha=0.0: L2 and L1 regularization terms on weights. We keep default values to maintain
  stability while avoiding unnecessary sparsity, as initial tests did not show significant imbalance issues.

- random_state=42: Ensures reproducibility of results across runs.

Overall, these hyperparameters provide a balanced starting point for XGBoost, ensuring good predictive power while
controlling overfitting.
"""
```
