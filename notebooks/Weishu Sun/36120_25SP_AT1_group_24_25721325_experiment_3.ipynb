```python
def split_dataset(X, y, test_size=0.2, random_state=42, stratify=True):
    """
    Parameters:
        X : DataFrame features
        y : Series labels
        test_size : float Test set proportion, default 0.2
        random_state : int Random seed for reproducibility
        stratify : bool Whether to perform stratified sampling by category

    return：
        X_train, X_valid, y_train, y_valid
    """
    if stratify:
        X_train, X_valid, y_train, y_valid = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )
    else:
        X_train, X_valid, y_train, y_valid = train_test_split(
            X, y, test_size=test_size, random_state=random_state
        )

    return X_train, X_valid, y_train, y_valid
```

```python
# <Student to fill this section>
def drop_leakage_features(X_train, X_test):
    """
    Remove features that may cause information leakage or overfitting
    """
    leakage_features = [
        "player_id_RARE", "Rec_Rank", "Rec_Rank__is_missing",

    ]

    features_to_drop = [col for col in leakage_features if col in X_train.columns]

    print("Dropping features:", features_to_drop)

    X_train_clean = X_train.drop(columns=features_to_drop, errors="ignore")
    X_test_clean = X_test.drop(columns=features_to_drop, errors="ignore")

    return X_train_clean, X_test_clean


X_train_final, X_test_final = drop_leakage_features(X_train_final, X_test_final)
```

```python
y_train = y_train_split
def align_columns(train_df, test_df, feature_list, fill_value=0):
    missing = [c for c in feature_list if c not in test_df.columns]
    for c in missing:
        test_df[c] = fill_value
    extra = [c for c in test_df.columns if c not in feature_list]
    if extra:
        test_df = test_df.drop(columns=extra)
    return train_df[feature_list], test_df[feature_list]

# Align the columns and obtain the final matrix to be saved.
X_train = X_train[features_list]
X_val   = X_valid[features_list]
X_test  = X_test_final.copy()

X_train, X_test = align_columns(X_train, X_test, features_list)
_,       X_val  = align_columns(X_train, X_val, features_list)

print("Train/Val/Test shapes:", X_train.shape, X_val.shape, X_test.shape)
```

### J.2 Set Hyperparameters

> Provide some explanations on why you believe this algorithm is a good fit


```python
pos = (y_train == 1).sum()
neg = (y_train == 0).sum()
scale_pos = neg / max(pos, 1)  # Positive class weight

print(f"Class balance: pos={pos}, neg={neg}, scale_pos_weight≈{scale_pos:.1f}")
```

```python

lgbm = LGBMClassifier(
    objective="binary",
    n_estimators=100,
    learning_rate=0.1,
    num_leaves=15,
    min_child_samples=10,
    max_bin=1023, min_data_in_bin=1,
    subsample=1.0, colsample_bytree=1.0,
    reg_alpha=0.0, reg_lambda=0.0,
    scale_pos_weight=1,
    random_state=42
)
```

```python
# <Student to fill this section>
hyperparameters_selection_explanations = """
In this experiment, we selected a relatively simplified and stable set of LightGBM hyperparameter configurations for the following primary reasons:

- n_estimators=100: Previous experiments revealed that excessively large numbers of trees often trigger the “no further splits with positive gain” warning, causing premature model training termination. Fixing n_estimators at 100 ensures stable training and prevents overfitting.
- learning_rate=0.1: Combined with a smaller number of trees, a learning rate of 0.1 enables rapid convergence within a finite number of iterations while preserving the model's generalization capability.
- num_leaves=15: A smaller number of leaves reduces tree complexity, preventing overfitting in the limited positive class samples while ensuring the model captures key nonlinear relationships.
- min_child_samples=10: This setting ensures each leaf contains a minimum number of samples to prevent overfitting. Its moderate value helps preserve effective splits for the minority class in imbalanced data.
- max_bin=1023, min_data_in_bin=1: Enhances feature binning precision to avoid information loss, suitable for datasets with numerous numerical features.
- subsample=1.0 and colsample_bytree=1.0: Disables random sampling of rows or columns, maintaining experimental control and stability.
- reg_alpha=0.0, reg_lambda=0.0: Additional regularization is disabled, ensuring the model runs with baseline settings to avoid overly restricting its capacity.
- scale_pos_weight=1: Despite class imbalance in the data, we use the default value of 1 in this baseline experiment to ensure model stability and prevent overfitting caused by excessively high weights. Recall and F1 scores will be optimized by adjusting weights and thresholds in subsequent experiments.

"""
```
